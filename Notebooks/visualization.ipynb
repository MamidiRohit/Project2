```python

import numpy as np
import matplotlib.pyplot as plt

def generate_data():
    np.random.seed(42)
    X = np.random.rand(100, 5)
    true_coeffs = np.array([2, -1, 3, 0, 0])  # True coefficients
    y = X @ true_coeffs + np.random.randn(100) * 0.5
    return X, y

def visualize_kfold_results(k_fold_mse_scores, k_fold_r2_scores):
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(k_fold_mse_scores) + 1), k_fold_mse_scores, marker='o', label='MSE')
    plt.plot(range(1, len(k_fold_r2_scores) + 1), k_fold_r2_scores, marker='s', label='R-Squared')
    plt.title('K-Fold Cross-Validation Results')
    plt.xlabel('Fold Number')
    plt.ylabel('Scores')
    plt.legend()
    plt.grid(True)
    plt.show()

def visualize_bootstrap_results(bootstrap_mse_scores, bootstrap_r2_scores):
    plt.figure(figsize=(8, 6))
    plt.plot(range(1, len(bootstrap_mse_scores) + 1), bootstrap_mse_scores, marker='o', label='MSE')
    plt.plot(range(1, len(bootstrap_r2_scores) + 1), bootstrap_r2_scores, marker='s', label='R-Squared')
    plt.title('Bootstrapping Results')
    plt.xlabel('Iteration Number')
    plt.ylabel('Scores')
    plt.legend()
    plt.grid(True)
    plt.show()

def visualize_comparison(k_fold_avg_mse, k_fold_avg_r2, bootstrap_avg_mse, bootstrap_avg_r2):
    methods = ['K-Fold', 'Bootstrapping']
    avg_mse = [k_fold_avg_mse, bootstrap_avg_mse]
    avg_r2 = [k_fold_avg_r2, bootstrap_avg_r2]

    plt.figure(figsize=(8, 6))
    plt.bar(methods, avg_mse, alpha=0.7, label='MSE', width=0.4, align='center')
    plt.bar(methods, avg_r2, alpha=0.7, label='R-Squared', width=0.4, align='edge')
    plt.title('Average Scores Comparison')
    plt.ylabel('Scores')
    plt.legend()
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # Simulate results
    X, y = generate_data()
    
    k_fold_mse_scores = [0.25, 0.24, 0.26, 0.23, 0.27]
    k_fold_r2_scores = [0.88, 0.89, 0.87, 0.90, 0.86]

    bootstrap_mse_scores = [0.25, 0.26, 0.24, 0.28, 0.23, 0.29, 0.27, 0.26, 0.24, 0.25]
    bootstrap_r2_scores = [0.88, 0.87, 0.89, 0.86, 0.90, 0.85, 0.87, 0.88, 0.89, 0.88]

    visualize_kfold_results(k_fold_mse_scores, k_fold_r2_scores)
    visualize_bootstrap_results(bootstrap_mse_scores, bootstrap_r2_scores)

    k_fold_avg_mse = np.mean(k_fold_mse_scores)
    k_fold_avg_r2 = np.mean(k_fold_r2_scores)
    bootstrap_avg_mse = np.mean(bootstrap_mse_scores)
    bootstrap_avg_r2 = np.mean(bootstrap_r2_scores)

    visualize_comparison(k_fold_avg_mse, k_fold_avg_r2, bootstrap_avg_mse, bootstrap_avg_r2)
```
