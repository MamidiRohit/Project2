Presentation and explanation of the Iris dataset

```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
data = load_iris()
X, y = data.data, data.target
print(X.shape)
print(y.shape)
```

```python
feature_names = data.feature_names
target_names = data.target_names

df = pd.DataFrame(X, columns=feature_names)
df['species'] = [target_names[i] for i in y]
```

```python
print("\nSummary Statistics:")
print(df.describe())
```

```python
sns.pairplot(df, hue='species', palette='Set1', markers=['o', 's', 'D'])
```

Note: The iris dataset has 150 samples, each with 4 features (sepal length, sepal width, petal length, petal width).

The dataset has 3 categories, setosa, versicolor, and virginica, each with 50 samples.
